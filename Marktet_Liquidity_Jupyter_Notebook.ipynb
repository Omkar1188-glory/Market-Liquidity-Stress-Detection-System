```python
df = df.sort_values("system_time")
```

🧠 Phase 4 – Liquidity Feature Engineering

🟢 STEP 1 — Spread-Based Liquidity Signals

```python
# Rolling mean & std (60-Minute window)
df["spread_mean_60"] = df["spread"].rolling(window=60).mean()
df["spread_std_60"] = df["spread"].rolling(window=60).std()

# z_score
df["spread_zscore"] = ( 
    (df["spread"] - df["spread_mean_60"])/ df["spread_std_60"]  )
```

```python
# spread_volatility
df["spread_volatility"] = df["spread"].rolling(window=60).std()
```

🟡 STEP 2 — Order Flow Imbalance

```python
# order flwo imbalance
df["flow_imbalance"] = ( 
    (df["buys"] - df["sells"]) / (df["buys"] + df["sells"])
)
```

🔵 STEP 3 — Depth Imbalance (Top 5 Levels)

```python
# Sum top 5 bid & ask limit depth
bid_depth = sum([df[f"bids_limit_notional_{i}"] for i in range(5)])
ask_depth = sum([df[f"asks_limit_notional_{i}"] for i in range(5)])

df["depth_imbalance"] = (
    (bid_depth - ask_depth) /
    (bid_depth + ask_depth)
)
```

🟠 STEP 4 — Cancellation Pressure

```python
bid_cancel = sum([df[f"bids_cancel_notional_{i}"] for i in range(5)])
ask_cancel = sum([df[f"asks_cancel_notional_{i}"] for i in range(5)])

bid_limit = sum([df[f"bids_limit_notional_{i}"] for i in range(5)])
ask_limit = sum([df[f"asks_limit_notional_{i}"] for i in range(5)])

df["cancel_pressure"]= (
    (bid_cancel + ask_cancel) / (bid_limit + ask_limit) 
)
```

🔴 STEP 5 — Market Aggression Ratio

```python
bid_market = sum([df[f"bids_market_notional_{i}"] for i in range(5)])
ask_market = sum([df[f"asks_market_notional_{i}"] for i in range(5)])

df["market_aggression"] = (
    (bid_market + ask_market) / (bid_limit + ask_limit) 
)
```

```python
df[[
    "spread_zscore",
    "flow_imbalance",
    "depth_imbalance",
    "cancel_pressure",
    "market_aggression"
]].describe()
```

🧠 Phase 5 – Build the Liquidity Stress Index (LSI)

🎯 Step 1 — Standardize All Signals

```python
# Normalize Signals

signals = [
    "spread_zscore",
    "flow_imbalance",
    "depth_imbalance",
    "cancel_pressure",
    "market_aggression"
]

for col in signals :
    df[col + "_norm"] = (
        (df[col] - df[col].mean()) / (df[col].std())
    )
```

🎯 Step 2 — Make Direction Consistent

```python
df["flow_imbalance_norm"] = df["flow_imbalance_norm"].abs()
df["depth_imbalance_norm"] = df["depth_imbalance_norm"].abs()
```

🎯 Step 3 — Build Composite Liquidity Stress Index

```python
df["LSI"] = (
    df["spread_zscore_norm"] +
    df["flow_imbalance_norm"] +
    df["depth_imbalance_norm"] +
    df["cancel_pressure_norm"] +
    df["market_aggression_norm"]
) / 5
```

🎯 Step 4 — Define Stress Regimes

```python
def classify_lsi(x):
    if x < 0.5:
        return "Normal"
    elif x < 1.5:
        return "Warning"
    else:
        return "Critical"

df["stress_regime"] = df["LSI"].apply(classify_lsi)
```
